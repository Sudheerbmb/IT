```python
import pandas as pd
import numpy as np
from scipy import stats

def process_column(file_name, column_name):
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_name)
    elif file_name.endswith('.xlsx'):
        df = pd.read_excel(file_name)
    else:
        raise ValueError("File format not supported. Please provide a CSV or Excel file.")

    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in the dataframe.")
    
    null_count = df[column_name].isnull().sum()
    if null_count > 0:
        print(f"Column '{column_name}' has {null_count} null values.")
        action = input("Would you like to remove null values (remove) or replace with mean, median, or mode? (mean/median/mode): ").strip().lower()
        if action == 'remove':
            df = df.dropna(subset=[column_name])
        elif action == 'mean':
            df[column_name].fillna(df[column_name].mean(), inplace=True)
        elif action == 'median':
            df[column_name].fillna(df[column_name].median(), inplace=True)
        elif action == 'mode':
            df[column_name].fillna(df[column_name].mode()[0], inplace=True)
        else:
            print("Invalid action. No changes made to null values.")
    
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound_iqr = Q1 - 1.5 * IQR
    upper_bound_iqr = Q3 + 1.5 * IQR
    outliers_iqr = df[(df[column_name] < lower_bound_iqr) | (df[column_name] > upper_bound_iqr)]
    print(f"\nIQR Method:\nNumber of outliers detected: {len(outliers_iqr)}")

    z_scores = stats.zscore(df[column_name].dropna())
    outliers_z = df[(np.abs(z_scores) > 3)]
    print(f"\nZ-score Method:\nNumber of outliers detected: {len(outliers_z)}")

    outlier_removal_method = input("\nWhich method would you like to use to remove outliers? (IQR/Z): ").strip().upper()
    if outlier_removal_method == 'IQR':
        df = df[(df[column_name] >= lower_bound_iqr) & (df[column_name] <= upper_bound_iqr)]
    elif outlier_removal_method == 'Z':
        df = df[np.abs(z_scores) <= 3]
    else:
        print("Invalid method. No changes made to outliers.")

    return df

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

def load_data(file_name):
    if file_name.endswith('.csv'):
        return pd.read_csv(file_name)
    elif file_name.endswith('.xlsx'):
        return pd.read_excel(file_name)
    else:
        raise ValueError("File format not supported. Please provide a CSV or Excel file.")

def display_statistics(df):
    print("\nData Overview:")
    print(df.describe(include='all'))
    print("\nNull Values:")
    print(df.isnull().sum())
    print("\nDuplicate Rows:")
    print(df.duplicated().sum())
    print("\nNegative Values:")
    negative_values = (df.select_dtypes(include=[np.number]) < 0).sum()
    print(negative_values[negative_values > 0])

def handle_null_values(df, column):
    action = input(f"Column '{column}' has null values. Would you like to remove (remove) or replace with mean, median, or mode? (mean/median/mode): ").strip().lower()
    if action == 'remove':
        df = df.dropna(subset=[column])
    elif action == 'mean':
        df[column].fillna(df[column].mean(), inplace=True)
    elif action == 'median':
        df[column].fillna(df[column].median(), inplace=True)
    elif action == 'mode':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        print("Invalid action. No changes made to null values.")
    return df

def handle_duplicates(df):
    action = input("Would you like to remove duplicate rows? (yes/no): ").strip().lower()
    if action == 'yes':
        df = df.drop_duplicates()
    return df

def handle_outliers(df, column):
    show_plots(df, column, "Before Outlier Removal")
    
    # Check for outliers using IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound_iqr = Q1 - 1.5 * IQR
    upper_bound_iqr = Q3 + 1.5 * IQR
    outliers_iqr = df[(df[column] < lower_bound_iqr) | (df[column] > upper_bound_iqr)]
    print(f"\nIQR Method:\nNumber of outliers detected: {len(outliers_iqr)}")

    z_scores = stats.zscore(df[column].dropna())
    outliers_z = df[(np.abs(z_scores) > 3)]
    print(f"\nZ-score Method:\nNumber of outliers detected: {len(outliers_z)}")

    outlier_removal_method = input("\nWhich method would you like to use to remove outliers? (IQR/Z): ").strip().upper()
    if outlier_removal_method == 'IQR':
        df = df[(df[column] >= lower_bound_iqr) & (df[column] <= upper_bound_iqr)]
    elif outlier_removal_method == 'Z':
        df = df[np.abs(z_scores) <= 3]
    else:
        print("Invalid method. No changes made to outliers.")
    
    show_plots(df, column, "After Outlier Removal")
    return df

def show_plots(df, column, title_suffix):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot - {column} ({title_suffix})')

    plt.subplot(1, 2, 2)
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram - {column} ({title_suffix})')
    plt.show()

def process_data(file_name):
    df = load_data(file_name)
    display_statistics(df)
    
    columns = input("Enter the column names you want to process (comma-separated) or type 'all' for all columns: ").strip().lower()
    if columns == 'all':
        columns = df.columns
    else:
        columns = [col.strip() for col in columns.split(',')]
    
    for column in columns:
        if df[column].isnull().sum() > 0:
            df = handle_null_values(df, column)
        
        if df.duplicated().sum() > 0:
            df = handle_duplicates(df)
        
        if pd.api.types.is_numeric_dtype(df[column]) and len(df[column].unique()) > 2: 
            df = handle_outliers(df, column)
    
    return df


df=process_data("campaign_performance.xlsx")
```

```python
import pandas as pd
from datetime import datetime

def extract_date_from_campaign(campaign_name):
    date_str = campaign_name[-6:]
    try:
        return datetime.strptime(date_str, '%d%m%y').strftime('%d-%m-%Y')
    except ValueError:
        return None

df['campaign_start_date'] = df.apply(
    lambda row: extract_date_from_campaign(row['campaign_name']) if pd.isnull(row['campaign_start_date']) else row['campaign_start_date'],
    axis=1
)

df['campaign_start_date'] = pd.to_datetime(df['campaign_start_date'], format='%d-%m-%Y', errors='coerce')

print(df)
```

```python
import pandas as pd

df['clicks'] = pd.to_numeric(df['clicks'], errors='coerce')
df['impressions'] = pd.to_numeric(df['impressions'], errors='coerce')

df['click_through_rate'] = df.apply(
    lambda row: (row['clicks'] / row['impressions'] * 100) if pd.isnull(row['click_through_rate']) and row['impressions'] > 0 else row['click_through_rate'],
    axis=1
)

df['click_through_rate'] = pd.to_numeric(df['click_through_rate'], errors='coerce')

print(df)
```

```python
def plot_donut(data, col):
    labels = data[col].value_counts().index
    sizes = data[col].value_counts().values
    colors = plt.get_cmap('viridis')(np.linspace(0, 1, len(labels)))
    
    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, colors=colors, startangle=90, wedgeprops=dict(width=0.3))
    plt.title(f'Donut Chart of {col}')
    plt.gca().set_aspect('equal')
    plt.show()

for col in categorical_cols:
    plot_donut(df, col)
```
