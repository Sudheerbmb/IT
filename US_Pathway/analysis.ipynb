```python
import pandas as pd
import numpy as np

# Example dataframe


# Select numerical columns for outlier calculation
numerical_columns = ['total_spent', 'impressions', 'clicks', 'click_through_rate', 'leads']

# Dictionary to store count of outliers for each column
outliers_count = {}

# Calculate outliers using Z-score method for each numerical column
for col in numerical_columns:
    # Calculate Z-score for each value in the column
    z_scores = np.abs((df[col] - df[col].mean()) / df[col].std())
    
    # Define threshold for outliers (e.g., Z-score > 3)
    threshold = 3
    
    # Count outliers
    outliers = df[z_scores > threshold]
    outliers_count[col] = len(outliers)

# Print count of outliers for each column
for col, count in outliers_count.items():
    print(f"Number of outliers in '{col}': {count}")
```

```python
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

numerical_columns = ['total_spent', 'impressions', 'clicks', 'click_through_rate', 'leads']


for col in numerical_columns:
    stat, p = stats.shapiro(df[col])
    print(f"Variable: {col}, Shapiro-Wilk test statistic: {stat:.4f}, p-value: {p:.4f}")
    
    # Visualize distribution
    plt.figure(figsize=(8, 4))
    plt.hist(df[col], bins=30, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```
